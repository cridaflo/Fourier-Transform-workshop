# Fourier series

Periodic functions can be represented a infinite sums of sines and cosines

$$f(x) = A_0 + \sum_{n=1}^{\infty} A_n \sin{\left(\frac{2\pi n x}{T}\right)} +  B_n \cos{\left(\frac{2\pi n x}{T}\right)}$$

Where
$$A_0 = \frac{1}{P}\int_{-T/2}^{T/2}f(x) \, dx$$
$$A_n = \frac{2}{P}\int_{-T/2}^{T/2}f(x) \cos{\left(\frac{2 \pi n x}{T}\right)} \, dx$$
$$B_n = \frac{2}{P}\int_{-T/2}^{T/2}f(x) \sin{\left(\frac{2 \pi n x}{T}\right)} \, dx$$

### Examples



```python
from math import pi
import numpy as np
from matplotlib import pyplot as plt

def fourier_sum(A0, As, Bs, T):
    x = np.linspace(0, 10, 1000)
    y = np.zeros(1000)
    for i in range(len(y)):
        y[i] = A0
        for n, a in enumerate(As):
            y[i] = y[i] + a*np.sin(2*pi*(n+1)*x[i]/T)
        for n, b in enumerate(Bs):
            y[i] = y[i] + b*np.cos(2*pi*(n+1)*x[i]/T)
    plt.plot(x, y)
    plt.show()
```

```python

def get_square_coefficients(n):
    ans = []
    for i in range(1, n+1):
        if i % 2 == 0:
            ans.append(0)
        else:
            val = 4/(pi * i)
            ans.append(val)
    return ans, [0]*n

n = 10
As, Bs = get_square_coefficients(n)
fourier_sum(0,As, Bs, 3)
```

```python
def get_triangle_coeficients(n):
    ans = []
    for i in range(1, n+1):
        if i%2 == 0:
            ans.append(0)
        else:
            val = 8/(pi*pi*i*i)
            ans.append(val)
    return [0]*n, ans

n = 100
As, Bs = get_triangle_coeficients(n)
fourier_sum(0,As, Bs, 3)
```

```python
n = 100
As, _ = get_square_coefficients(n)
_ , Bs = get_triangle_coeficients(n)
fourier_sum(0,As, Bs, 3)
```

## Exponential form

Expressing function in terms of sines and cosines is not very practical


We can leverage the folling equations:
$$ e^{ix} = \cos{(x)} + i\sin{(x)}$$



And get
$$f(x) =\sum_{n=-\infty}^{\infty} c_n \cdot e^{\frac{2\pi i n x}{T}}$$
Were 
$$ c_n = \frac{1}{P} \int_{-P/2}^{P/2} f(x) e^{\frac{2\pi i n x}{T}} \, dx$$
    

### Examples

```python
import cmath
i = complex(0, 1)

def generate_complex_coefficients(A0, As, Bs, maxi):
    cs = np.zeros(2*maxi+1, dtype=complex)
    cs[0] = A0
    for n in range(maxi):
        cs[n+1] = complex(Bs[n]/2, -As[n]/2)
        cs[-n-1] = complex(Bs[n]/2, As[n]/2)
    return cs

def complex_fourier_sum(cs, T, maxi):
    x = np.linspace(0, 10, 1000)
    y = np.zeros(1000, dtype=complex)
    for pos in range(len(x)):
        for n in range(-maxi, maxi+1):
            y[pos] = y[pos] + cs[n]*cmath.exp(2*pi*i*n*x[pos]/T)
    plt.plot(x, y)
    plt.show()
```

```python
n = 8
As, Bs = get_square_coefficients(n)
cs = generate_complex_coefficients(0, As, Bs, n)
complex_fourier_sum(cs, 3, n)
```

```python
n = 6
As, Bs = get_triangle_coeficients(n)
cs = generate_complex_coefficients(0, As, Bs, n)
complex_fourier_sum(cs, 3, n)
```

# Fourier Transform

What if the the function is not periodic?

In this case we have a continium of frecuencies

$$\mathscr{F}\{f\} = \hat{f}(\xi) =  \int_{-\infty}^{\infty} f(x) e^{-2\pi i\xi x} \, dx$$

Fuction $\hat{f}$ is said to be in the frecuency domain

The transformation is invertible. Thus, the original function can be recovered from by the following relation
$$\mathscr{F}^{-1}\{\hat{f}\} = f(x) =  \int_{-\infty}^{\infty} \hat{f}(x) e^{2\pi i\xi x} \, d\xi$$

## Example 

```python
from scipy.io import wavfile
import scipy.io
from scipy.fft import fft, fftfreq
from pydub import AudioSegment
from pydub.playback import play
```

```python
wav_fname = 'example_1.wav'
samplerate, data = wavfile.read(wav_fname)
length = data.shape[0] / samplerate
time = np.linspace(0., length, data.shape[0])
plt.plot(time, data)
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.show()
```

```python
plt.plot(time[:100], data[:100])
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.show()
```

```python
yf = fft(data)
xf = fftfreq(data.shape[0], 1 / samplerate)
plt.plot(xf, np.abs(yf))
plt.show()

plt.plot(xf, np.angle(yf))
plt.show()
```

# Discrete Fourier Transform

In real life we usually don't deal with with analytic functions.

Instead we have finite arrays of sampled values.

We use a related transformation which takes these values a produces a sequence of samples in frequency domain.



Given a list of equaly spaced samples ${x_n}_{n \in N}$ we get a sequence ${X_k}_{k \in N}$ of equally spaced samples in frecuency domain. 

$$X_k = \sum_{n=0}^{N-1} x_n \cdot e^{-\frac{2\pi i kn}{N}}$$


The distance between these samples is the inverse of the samplig frequency of the original sequence.

Once again, this operation is invertible.

We can recover the original sequence by the folllowing operation:

$$x_n = \sum_{k = 0}^{N-1} X_k \cdot e^{\frac{2 \pi i k n}{N}}$$

This is the operation that is actually performed by computers when we refer to fourier transform.

The algorithm used to calculate this efficiently is called Fast Fourier Transform (FFT).
